```python
def mlplot(
    x,
    y,
    title="", 
    annotation=None,
    anotx=40,
    anoty=42,
    anoax=20,
    anoay=80,
    anosize='x-small',
    xlab="input",
    ylab="output",
    size=(8,7),
    filename=None,
    logox=98,
    logoy=-4
    
):
    plt.figure(facecolor='#000000',edgecolor='#000000',figsize=size,layout='tight')
    plt.axes(facecolor='#000000')
    plt.tick_params(colors='white', grid_color='white')
    plt.xticks(fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold')
    plt.yticks(fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold')
    plt.title(title, fontfamily='monospace', fontsize='medium', fontstyle='italic', fontweight='semibold', color='#ffffff')
    if annotation is not None:
        plt.annotate(annotation,(anotx, anoty), (anoax,anoay), arrowprops={'arrowstyle':'-|>','color':'#ffffff'}, color='#ffffff', fontfamily='monospace', fontsize=anosize,fontstyle='italic', fontweight='semibold')
    plt.xlabel(xlab, loc='left', fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold', color='#ffffff')
    plt.ylabel(ylab, loc='bottom', fontfamily='monospace', fontsize='small', fontstyle='italic', fontweight='semibold', color='#ffffff')
    mpl.rc('axes', edgecolor='white', facecolor='white')
    plt.text(logox,logoy,'machine-labs.io',fontfamily='monospace', fontsize='xx-small',fontstyle='italic', fontweight='semibold', color='#ffffff')
    plt.plot(x,y,color='#ffffff');
    
    if filename is not None:
        plt.savefig(filename, metadata={'title':f'{filename}','author':'machine-labs.io'})


# O(n)
def lin(n):
    _ = 0
    for i in range(n):
        _+=1
    return (n,_)

x=[]
y=[]
for i in range(100):
    x.append(lin(i)[0])
    y.append(lin(i)[1])

mlplot(x,y,
        annotation='As the input size increases,\nthe output rate increases,\nat the same rate.',
        title='Linear Complexity, $O(n)$',
        size=(8,7),
        logox=90,
        filename='linear',
        xlab='The length of an input.',
        ylab='The rate of output growth.' # check my language here
      )
```

```python
def mlplot(
    x,
    y,
    title="", 
    annotation=None,
    anotx=40,
    anoty=2000,
    anoax=10,
    anoay=8500,
    anosize='x-small',
    xlab="input",
    ylab="output",
    size=(8,7),
    filename=None,
    logox=92,
    logoy=-400
    
):
    plt.figure(facecolor='#000000',edgecolor='#000000',figsize=size,layout='tight')
    plt.axes(facecolor='#000000')
    plt.tick_params(colors='white', grid_color='white')
    plt.xticks(fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold')
    plt.yticks(fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold')
    plt.title(title, fontfamily='monospace', fontsize='medium', fontstyle='italic', fontweight='semibold', color='#ffffff')
    if annotation is not None:
        plt.annotate(annotation,(anotx, anoty), (anoax,anoay), arrowprops={'arrowstyle':'-|>','color':'#ffffff'}, color='#ffffff', fontfamily='monospace', fontsize=anosize,fontstyle='italic', fontweight='semibold')
    plt.xlabel(xlab, loc='left', fontfamily='monospace', fontsize='small',fontstyle='italic', fontweight='semibold', color='#ffffff')
    plt.ylabel(ylab, loc='bottom', fontfamily='monospace', fontsize='small', fontstyle='italic', fontweight='semibold', color='#ffffff')
    mpl.rc('axes', edgecolor='white', facecolor='white')
    plt.text(logox,logoy,'machine-labs.io',fontfamily='monospace', fontsize='xx-small',fontstyle='italic', fontweight='semibold', color='#ffffff')
    plt.plot(x,y,color='#ffffff');
    
    if filename is not None:
        plt.savefig(filename, metadata={'title':f'{filename}','author':'machine-labs.io'})


# O(n^2) generate input
# two nested for loops
# that take n as the input
# means we process n*n
# values. n^2.
def square(n):
    y = 0
    for i in range(n):       # these two for loops take
        for j in range(n):   # an input as the range
            y+=1
    return (n,y)

x=[]
y=[]

for i in range(100):
    x.append(square(i)[0])
    y.append(square(i)[1])

mlplot(x,y,
        annotation='As the input size increases,\nthe output rate increases,\nat the input rate squared.',
        title='Quadratic Complexity, $O(n^2)$',
        size=(8,7),
        logox=90,
        filename='quadratic',
        xlab='The length of an array.',
        ylab='The rate of output growth.' # check my language here
      )
```
