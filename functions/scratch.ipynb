```python
import numpy as np
import matplotlib.pyplot as plt
import scipy


alpha = 2
_beta = 500

beta_dist = scipy.stats.beta(alpha, _beta)

x = np.linspace(0, 1, 1000)
y = beta_dist.pdf(x, alpha, _beta)

plt.plot(x, y)
plt.title('Beta Distribution (α=2, β=5)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
